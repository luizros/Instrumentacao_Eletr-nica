```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Lendo os dados do arquivo CSV
df = pd.read_csv('./giro.csv', delimiter=';', header=None)

# Separando os ângulos reais e os valores estimados
angulos_reais = df.iloc[0, :].values
valores_estimados = df.iloc[1:, :].values

angulos_reais
```

```python
rmse_values = []

for i in range(len(angulos_reais)):
    diferenca_quadratica = np.square(valores_estimados[:, i] - angulos_reais[i])
    mse = np.mean(diferenca_quadratica)
    rmse = np.sqrt(mse)
    rmse_values.append(rmse)
print("Ângulo \t\t|\t REMQ ")
for i, rmse_valor in enumerate(rmse_values):
    print(angulos_reais[i], "\t\t:\t" , rmse_valor)
```

```python
plt.plot(angulos_reais, rmse_values, marker='o')
plt.title('RMSE em Função da Variação Angular')
plt.xlabel('Variação Angular (graus)')
plt.ylabel('RMSE')
plt.show()
```
